```python
def pcc(X, Y):
   # Normalise X and Y
    X -= X.mean(0)
    Y -= Y.mean(0)
   # Standardise X and Y
    X /= X.std(0)
    Y /= Y.std(0)
   # Compute mean product
    return np.mean(X*Y)
```
