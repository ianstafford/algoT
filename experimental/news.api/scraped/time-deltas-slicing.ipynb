```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Timedelta
import numpy as np
%matplotlib inline
sns.set()
```

```python
freq = Timedelta(minutes=5)
```

```python
df = pd.read_csv('intraday_5min_SBIN.csv', parse_dates=['timestamp']).sort_values('timestamp').set_index('timestamp')
df.index = df.index + Timedelta(hours=9, minutes=30) # Converting to Indian Time
```

```python
df['change'] = df['close'].diff()
```

```python
df['2018-6-8']
```

```python
slice = df['2018-6-8']
```

```python
for index, row in slice.iterrows():
    print(slice.loc[index])
```

```python
window = 13
for index, row in slice.iterrows():
    if window > 0:
        window -= 1
        continue
    print(slice.loc[index-(13*freq):index])
```
